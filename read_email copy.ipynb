```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

class GmailAutomation:
    def __init__(self, email_address, password) -> None:
        self.email_address = email_address
        self.password = password
        self.driver = webdriver.Chrome()

    def login(self):
        self.driver.get("https://mail.google.com/")
        email_input = self.driver.find_element("name", "identifier")
        email_input.send_keys(self.email_address)
        email_input.send_keys(Keys.RETURN)
        time.sleep(2)  # Wait for the password input to be visible
        password_input = self.driver.find_element("name", "password")
        password_input.send_keys(self.password)
        password_input.send_keys(Keys.RETURN)
        time.sleep(5)  # Wait for the inbox to load

    def open_email_by_id(self, email_id):
        self.driver.get(f"https://mail.google.com/mail/u/0/#inbox/{email_id}")

    def extract_email_details(self):
        subject = self.driver.find_element_by_xpath('//h2[@class="hP"]').text
        from_address = self.driver.find_element_by_xpath('//span[@class="go"]').text
        date_sent = self.driver.find_element_by_xpath('//span[@class="g3"]').text

        return {"Subject": subject, "From": from_address, "Date": date_sent}

    def close_browser(self):
        self.driver.quit()

# Example usage
gmail_automation = GmailAutomation(email_address='phtrieu1040@gmail.com', password='kxwi hwjj tbdu qgtg')
gmail_automation.login()
gmail_automation.open_email_by_id('10065')
email_details = gmail_automation.extract_email_details()
print(email_details)
gmail_automation.close_browser()
```
